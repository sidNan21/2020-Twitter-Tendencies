# Comparison of the Average Number of Favorites and Retweets per Tweet for the Last 3000 Tweets by the 2016 Presidential Nominees and Top Prospective 2020 Democratic Presidential Candidates

```python
# Package imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import twitter
```

```python
# Twitter Handles for Each Politician

# 2016 Nominees
trump = 'realDonaldTrump'
clinton = 'HillaryClinton'

# Top 6 Prospects
biden = 'JoeBiden'
booker = 'CoryBooker'
sanders = 'BernieSanders'

harris = 'KamalaHarris'
gillibrand = 'SenGillibrand'
warren = 'elizabethforma'
```

```python
# Data Frames for Favorites/Retweets/Tweets for Each Politician
trump_data = twitter.tweets_favorites_retweets(trump)
clinton_data = twitter.tweets_favorites_retweets(clinton)

biden_data = twitter.tweets_favorites_retweets(biden)
booker_data = twitter.tweets_favorites_retweets(booker)
sanders_data = twitter.tweets_favorites_retweets(sanders)

harris_data = twitter.tweets_favorites_retweets(harris)
gillibrand_data = twitter.tweets_favorites_retweets(gillibrand)
warren_data = twitter.tweets_favorites_retweets(warren)
```

```python
# Creates various Data Series and combines them into two Data Frames for later plotting
nominees = pd.Series(['President Donald J. Trump', 'President Donald J. Trump', 'Hillary Clinton', "Hillary Clinton"])
nums = pd.Series([trump_data['favorites'].mean(), trump_data['retweets'].mean(),
                  clinton_data['favorites'].mean(), clinton_data['retweets'].mean()])
variables = pd.Series(['Favorited', 'Retweeted', 'Favorited', 'Retweeted'])

nominee_data = pd.DataFrame({'Politician': nominees, 'Action on Tweet': variables, 'Average Number of People': nums})

prospects = pd.Series(['Joe Biden', 'Joe Biden', 'Cory Booker', 'Cory Booker', 'Bernie Sanders',
                       'Bernie Sanders', 'Kamala Harris', 'Kamala Harris', 'Kristen Gillibrand',
                       'Kristen Gillibrand', 'Elizabeth Warren', 'Elizabeth Warren'])
prospect_nums = pd.Series([biden_data['favorites'].mean(), biden_data['retweets'].mean(),
                           booker_data['favorites'].mean(), booker_data['retweets'].mean(),
                           sanders_data['favorites'].mean(), sanders_data['retweets'].mean(),
                           harris_data['favorites'].mean(), harris_data['retweets'].mean(),
                           gillibrand_data['favorites'].mean(), gillibrand_data['retweets'].mean(),
                           warren_data['favorites'].mean(), warren_data['retweets'].mean()])
variables_two = pd.Series(['Favorited', 'Retweeted', 'Favorited', 'Retweeted',
                           'Favorited', 'Retweeted', 'Favorited', 'Retweeted',
                           'Favorited', 'Retweeted', 'Favorited', 'Retweeted'])

prospect_data = pd.DataFrame({'Politician': prospects, 'Action on Tweet': variables_two,
                              'Average Number of People': prospect_nums})
```

```python
# Creates figure comparing 2016 Presidential Nominees
f, (forty_five) = plt.subplots(figsize=(16, 8))
sns.set_style("whitegrid")
rb = sns.diverging_palette(10, 220, sep=80, n=2)
forty_five.set_title('Comparison of Average Number of Favorites and Retweets of the Last 3000 Tweets by President Trump or Hillary Clinton',
                     fontstyle='italic', fontsize=16)

forty_five = sns.barplot(x="Politician", y ='Average Number of People', hue='Action on Tweet',
                         palette=rb, data=nominee_data)
```

```python
# Creates figure comparing Top 6 Potential 2020 Democratic Presidential Candidates
f, (twenty_twenty) = plt.subplots(figsize=(16, 8))
sns.set_style("whitegrid")
rb = sns.diverging_palette(10, 220, sep=80, n=2)
twenty_twenty.set_title('Comparison of Average Number of Favorites and Retweets of the Last 3000 Tweets between the Top 6 Potential 2020 Democratic Candidates',
                        fontstyle='italic', fontsize=14)

twenty_twenty = sns.barplot(x="Politician", y ='Average Number of People', hue='Action on Tweet', 
                            palette=rb, data=prospect_data)
```

A visual representation of the amount of average favorites and retweets per tweet from each of these politicians illustrates the overwhelming online presence of Donald Trump's twitter account. The average number of favorites on one of President Trump's tweets is slightly higher thatn 80,000 while Hillary Clinton manages only an average of around 20,000 and Bernie Sanders takes a distant third place at nearly half of Hillary Clinton's average at around 9,000. The same rankings are clear in the average rewtweet count with Trump taking an overwhelming lead at an average of nearly 20K retweets per tweet, Clinton second at less than half of Trump's average, and Sanders at third with around half of Clinton's average.
